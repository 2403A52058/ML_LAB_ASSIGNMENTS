```python
df['Math_Result'] = df['math_score'].apply(lambda x: 'Above 70' if x > 70 else '70 or Below')
df['Math_Result'].value_counts()
```

```python
# Select score columns
scores_df = df[['math_score', 'science_score', 'english_score']]

# Convert to NumPy array
scores_np = scores_df.to_numpy()

scores_np
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['math_score'], bins=10)
plt.xlabel('Math Scores')
plt.ylabel('Number of Students')
plt.title('Histogram of Math Scores')
plt.show()
```

```python
subjects = ['math_score', 'science_score', 'english_score']
averages = [df[sub].mean() for sub in subjects]

plt.figure()
plt.bar(['Math', 'Science', 'English'], averages)
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.title('Average Scores by Subject')
plt.show()
```

```python
plt.figure()
plt.scatter(df['math_score'], df['english_score'])
plt.xlabel('Math Scores')
plt.ylabel('English (Writing) Scores')
plt.title('Math vs Writing (English) Scores')
plt.show()
```

```python
gender_group = df.groupby('gender')[['math_score', 'science_score', 'english_score']].mean()
gender_group
```

```python
gender_group.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Average Scores by Gender')
plt.show()
```

```python
parent_group = df.groupby('parent_education')[['math_score', 'science_score', 'english_score']].mean()
parent_group
```

```python
parent_group.plot(kind='bar')
plt.xlabel('Parental Education')
plt.ylabel('Average Score')
plt.title('Average Scores by Parental Education')
plt.xticks(rotation=45)
plt.show()
```
